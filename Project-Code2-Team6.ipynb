```python
import pandas as pd
import numpy as np
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model 
from sklearn.ensemble import RandomForestClassifier
import sklearn
from sklearn.metrics import log_loss 
```

```python
# Read in the data 
train = pd.read_csv('C:/Users/kasan/OneDrive/Desktop/Machine Learning/Final Project Datasets/newTrainingDataset.csv')
val = pd.read_csv('C:/Users/kasan/OneDrive/Desktop/Machine Learning/Final Project Datasets/newValidationDataset.csv')

train.head()
```

```python
trainX = train.iloc[:,1:]
trainY = train.iloc[:,0]

trainY.head()
```

```python
trainX.head()
```

```python
valX = val.iloc[:,1:]
valY = val.iloc[:,0]
```

# Random Forest 

```python
# Model 1 - before hyperparameter tuning
clf = RandomForestClassifier(max_depth=2, random_state=42)

# Fit the model 
clf = clf.fit(trainX, trainY) 
```

```python
pred_prob = clf.fit(trainX, trainY).predict_proba(valX) 
```

```python
sklearn.metrics.log_loss(valY, pred_prob, normalize=True, sample_weight=None, labels=None)
```

```python
from sklearn.model_selection import KFold
inner_cv=KFold(n_splits = 5, shuffle = True)

# gs for n_estimators parameter 
gs_rf = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                  param_grid=[{'n_estimators': [10,100,200,500,700,1000]}],
                  scoring='neg_log_loss',
                  cv=inner_cv)

gs_rf = gs_rf.fit(trainX,trainY)
print("\n Parameter Tuning") 
print("Optimal Parameter: ", gs_rf.best_params_)
print("Optimal Estimator: ", gs_rf.best_estimator_)
```

```python
# gs for min_samples_split parameter, put in optimal value for n_estimators
inner_cv=KFold(n_splits = 5, shuffle = True)

gs_rf2 = GridSearchCV(estimator=RandomForestClassifier(random_state=42, n_estimators = ),
                  param_grid=[{'min_samples_split': [2,4,6,8,10]}],
                  scoring='neg_log_loss',
                  cv=inner_cv)

gs_rf2 = gs_rf2.fit(trainX,trainY)
print("\n Parameter Tuning") 
print("Optimal Parameter: ", gs_rf2.best_params_)
print("Optimal Estimator: ", gs_rf2.best_estimator_)
```
